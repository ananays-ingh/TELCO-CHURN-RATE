```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/mr.engineer/Desktop/Code/DSPROJECT/CUSTOMER_CHURN/DATA/Telco_customer_churn.csv')
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data= df, x='Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.title('Label Distribution')
```

```python
columns = df.drop(columns=['customerID','TotalCharges',]).columns
plt.figure(figsize=(30,60))
j = 1
for i in range(18):
    if(columns[i]=='tenure' or columns[i]=='MonthlyCharges'):
        continue
    plt.subplot(8,2,j)
    sns.countplot(data=df,x = df[columns[i]],hue = 'Churn')
    plt.xlabel(columns[i])
    plt.ylabel("Count")
    plt.title("variation of Churn According to:" + columns[i])
    plt.grid(True)
    j+=1
plt.show()
```

```python
invalid_rows = df[pd.to_numeric(df['TotalCharges'], errors='coerce').isna()]

print("Number of problematic rows:", invalid_rows.shape[0])
print(invalid_rows[['customerID', 'TotalCharges', 'tenure', 'MonthlyCharges']])
```

```python
plt.figure(figsize= (15,5))

plt.subplot(1,3,1)
sns.boxplot(df['MonthlyCharges'])
plt.xlabel("MonthlyCharges")
plt.ylabel("Count")
plt.subplot(1,3,2)
sns.boxplot(df['tenure'])
plt.xlabel("tenure")
plt.ylabel("Count")
plt.subplot(1,3,3)
sns.boxplot(pd.to_numeric(df['TotalCharges'],errors = 'coerce'))
plt.xlabel("TotalCharges")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df[df["Churn"] == "No"]["tenure"], label="No Churn", fill=True)
sns.kdeplot(df[df["Churn"] == "Yes"]["tenure"], label="Churn", fill=True)
plt.title("Tenure Distribution by Churn")
plt.xlabel("Tenure")
plt.legend()
plt.show()
```
